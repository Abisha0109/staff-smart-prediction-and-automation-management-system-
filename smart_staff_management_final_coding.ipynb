```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/final_dataset_no_null_replacements.xlsx")
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab=LabelEncoder()
df['department']=lab.fit_transform(df['department'])
df['day_of_week']=lab.fit_transform(df['day_of_week'])
df['replacement_status']=lab.fit_transform(df['replacement_status'])
df['alert_status']=lab.fit_transform(df['alert_status'])
df['replacement_staff_id']=lab.fit_transform(df['replacement_staff_id'])
df['staff_id']=lab.fit_transform(df['staff_id'])
df['subject']=lab.fit_transform(df['subject'])
df['class_name']=lab.fit_transform(df['class_name'])
df
```

```python
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

    df["Year"] = df["date"].dt.year
    df["Day"] = df["date"].dt.day
    df["Month"] = df["date"].dt.month

    df.drop(columns=["date"], inplace=True)
    print(df.head())
```

```python
if "end_time" in df.columns:
    df["end_time"] = pd.to_datetime(df["end_time"], errors="coerce")

    df["Time_hour"] = df["end_time"].dt.hour
    df["Time_minute"] = df["end_time"].dt.minute
    df["Time_second"] = df["end_time"].dt.second

    df.drop(columns=["end_time"], inplace=True)
    print(df.head())
```

```python
if "start_time" in df.columns:
    df["start_time"] = pd.to_datetime(df["start_time"], errors="coerce")

    df["Time_hour"] = df["start_time"].dt.hour
    df["Time_minute"] = df["start_time"].dt.minute
    df["Time_second"] = df["start_time"].dt.second

    df.drop(columns=["start_time"], inplace=True)
    print(df.head())
```

```python
x=df.drop(['replacement_status'],axis=1)
y=df['replacement_staff_id']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(
    n_estimators=100,
    max_depth=3,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42
)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc=accuracy_score(y_test, y_pred)
print(acc)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=4,
    random_state=42
)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

```python
import datetime
```

```python
replacement_data = [
    {"absent_staff":101, "replacement_staff":205, "period":1, "time":"09:00"},
    {"absent_staff":102, "replacement_staff":210, "period":2, "time":"10:00"}
]
```

```python
def convert_to_datetime(time_str):
    today = datetime.date.today()
    hour, minute = map(int, time_str.split(":"))
    return datetime.datetime(today.year, today.month, today.day, hour, minute)
```

```python
def send_notification(replacement_data):

    current_time = datetime.datetime.now()

    for record in replacement_data:

        period_time = convert_to_datetime(record["time"])
        time_difference = (period_time - current_time).total_seconds() / 60

        # alert 15 minutes before period
        if 0 < time_difference <= 15:

            print("🔔 ALERT!")
            print(f"Replacement Staff {record['replacement_staff']} assigned")
            print(f"Period: {record['period']}")
            print(f"Time: {record['time']}")
            print("--------------------------")
send_notification(replacement_data)
```

```python
send_notification(replacement_data)
print(replacement_data)
print("🔔 ALERT!")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np

# -------------------- PAGE CONFIG --------------------
# -------------------- PAGE CONFIG --------------------
st.set_page_config(
    page_title="Smart Staff Management",
    page_icon="📊",
    layout="centered"
)


# -------------------- SIDEBAR --------------------
st.sidebar.title("📊 Smart Staff System")

menu = st.sidebar.radio(
    "Navigation",
    ["Dashboard", "Staff Prediction", "Replacement Staff", "Notifications"]
)

st.sidebar.markdown("---")
st.sidebar.info("AI-powered staff availability & replacement system")

# -------------------- SAMPLE DATA --------------------
data = {
    "Staff ID": [101, 102, 103, 104, 105],
    "Name": ["Arun", "Divya", "Kumar", "Priya", "Rahul"],
    "Status": ["Present", "Absent", "Present", "Absent", "Present"],
    "Department": ["IT", "HR", "IT", "Finance", "Admin"]
}

df = pd.DataFrame(data)

# -------------------- DASHBOARD --------------------
if menu == "Dashboard":

    st.title("📊 Smart Staff Management Dashboard")

    col1, col2, col3 = st.columns(3)

    total_staff = len(df)
    absent_today = len(df[df["Status"] == "Absent"])
    replacements = 2

    col1.metric("👥 Total Staff", total_staff)
    col2.metric("❌ Absent Today", absent_today)
    col3.metric("🔁 Replacements Assigned", replacements)

    st.markdown("---")

    st.subheader("📈 Staff Availability Overview")
    status_count = df["Status"].value_counts()
    st.bar_chart(status_count)

    st.subheader("📊 Department Distribution")
    dept_count = df["Department"].value_counts()
    st.line_chart(dept_count)

# -------------------- STAFF PREDICTION --------------------
elif menu == "Staff Prediction":

    st.title("🤖 Staff Availability Prediction")

    st.markdown("Enter Staff ID to predict availability")

    # TEXT INPUT → NO + − symbols
    staff_id = st.text_input("Enter Staff ID", placeholder="Type staff ID here...")

    if st.button("Predict Availability"):

        if staff_id == "":
            st.warning("Please enter Staff ID")

        else:
            prediction = np.random.choice(["Available", "Absent"])

            if prediction == "Available":
                st.success(f"Staff {staff_id} is AVAILABLE for the period")
            else:
                st.error(f"Staff {staff_id} is ABSENT — Replacement required")

# -------------------- REPLACEMENT STAFF --------------------
elif menu == "Replacement Staff":

    st.title("🔁 Replacement Staff Assignment")

    absent_staff = df[df["Status"] == "Absent"]

    st.subheader("Absent Staff List")
    st.dataframe(absent_staff)

    st.subheader("Assign Replacement Staff")

    replacement = st.selectbox("Select Replacement", df["Name"])

    if st.button("Assign Replacement"):
        st.success(f"{replacement} assigned successfully!")

# -------------------- NOTIFICATIONS --------------------
elif menu == "Notifications":

    st.title("🔔 Notification Alert System")

    message = st.text_area("Enter Notification Message")

    if st.button("Send Notification"):
        st.success("Notification sent successfully!")

    st.subheader("Recent Alerts")

    alerts = pd.DataFrame({
        "Time": ["9:00 AM", "10:30 AM"],
        "Message": ["Staff 102 absent", "Replacement assigned"]
    })

    st.table(alerts)
```
